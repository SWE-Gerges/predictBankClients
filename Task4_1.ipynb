```python
# importing used librares
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
import tensorflow.python.keras
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense
```

```python
data = pd.read_csv('bank_clients.csv')
data_CPY = data
```

```python
data.head(10)
```

```python
data = data.drop('CustomerId', axis =1)
data = data.drop('Surname', axis =1)
data = data.drop('RowNumber', axis =1)

data.head()
```

```python
data.isna().sum()
```

```python
dummies = pd.get_dummies(data['Geography'])
```

```python
dummies.head()
```

```python
data = pd.concat([data, dummies], axis =1)
```

```python
data.head()
```

```python
data = data.drop('Geography', axis = 1)
data = data.drop('Germany', axis =1)
```

```python
dummies = pd.get_dummies(data['Gender'], prefix = "Gender")
data = pd.concat([data, dummies], axis =1)
```

```python
data = data.drop('Gender_Male', axis =1)
data = data.drop('Gender', axis =1)
```

```python
data.head()
```

```python
y =  data['Exited']
X = data[['CreditScore', 'Age','Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary']]
print(y.head())
print(X.head())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
X.columns
```

```python
sc = StandardScaler()
X = pd.DataFrame(sc.fit_transform(X), columns = ['CreditScore', 'Age','Tenure', 'Balance', 'NumOfProducts', 'EstimatedSalary'])
```

```python
X = pd.concat([X, data[['HasCrCard', 'IsActiveMember', 'France', 'Spain', 'Gender_Female']]], axis = 1)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model = Sequential()
```

```python
model.add(Dense(6, input_dim =11, activation = 'relu'))
model.add(Dense(5, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'mean_squared_error', optimizer = 'adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs = 100)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_test
```

```python
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
```

```python
print(accuracy_score(pred, y_test))
```
